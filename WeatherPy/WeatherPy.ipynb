```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Print log of each city, with name, number and set
# Sorting the cities in alphabetical order for a better view

url  = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
cities = sorted(cities)
c = 0
s = 1
print("Beginning Data Retrieval")
print("-----------------------------------------------")
for city in cities:
    link =  url + "appid=" + weather_api_key + "&units="+units+"&q=" + city
    response = requests.get(link)
    try:
        if c == 50:
            c = 0
            s = s + 1
            weather_json = response.json()
            print(f"Processing Record {c} of Set {s} | {weather_json['name']}.")
            c = c + 1
        else:
            weather_json = response.json()
            print(f"Processing Record {c} of Set {s} | {weather_json['name']}.")
            c = c + 1       
    except:
        print("City not found. Skipping...")
print("-----------------------------------------------") 
print("Data Retrieval Complete")
print("-----------------------------------------------") 
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Converting the data retrieved into a Dataframe

url  = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
city_names = []
lat_list = []
lng_list = []
max_temp = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
country_names = []
dates_list = []

cities = sorted(cities)

c = 0
for city in cities:
    link =  url + "appid=" + weather_api_key + "&units="+units+"&q=" + city
    response = requests.get(link)
    try:
        weather_json = response.json()
        city_names.append(weather_json['name'])
        lat_list.append(weather_json['coord']['lat'])
        lng_list.append(weather_json['coord']['lon'])
        max_temp.append(weather_json['main']['temp_max'])
        humidity_list.append(weather_json['main']['humidity'])
        cloudiness_list.append(weather_json['clouds']['all'])
        wind_speed_list.append(weather_json['wind']['speed'])
        country_names.append(weather_json['sys']['country'])
        dates_list.append(weather_json['dt'])
        
    except:
        pass
weather_df  = pd.DataFrame(list(zip(city_names,lat_list,lng_list,max_temp,humidity_list,
                                   cloudiness_list,wind_speed_list,country_names,dates_list)),
                           columns=["City","Latitude","Lng","Max Temperature","Humidity","Cloudiness","Wind Speed","Country","Date"]
                          )
weather_df.head()
```

```python
# Exporting the city data into a csv file
weather_df.to_csv("WeatherPy_Outputs/My_Cities_Data.csv",index=False)
```

```python
# Converting Unix Epoch Time To Datetime
date_value = weather_df["Date"].mean()
actual_date = pd.to_datetime(date_value, unit = 's')
real_date = actual_date.date()
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
# Looking for humid cities (humidity > 100 %)

humid_cities = weather_df.loc[weather_df["Humidity"]>100,:]
humid_cities
```

```python
#  Get the indices of cities that have humidity over 100%.

humid_cities_index = humid_cities.index
humid_cities_index_list = humid_cities_index.to_list()
```

```python
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.
# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".

weather_df = weather_df.drop(humid_cities_index_list)
weather_df.head()
```

## Linear Regression

```python
northern_hemisphere_df = weather_df.loc[weather_df["Latitude"] >= 0,:]
southern_hemisphere_df = weather_df.loc[weather_df["Latitude"] < 0,:]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.figure(figsize=(7,5))

x_values = northern_hemisphere_df["Latitude"]
y_values = northern_hemisphere_df["Max Temperature"]

(slope,intercept,rvalue,pvalue,stderr) = stats.linregress(x_values,y_values)
regress_values = x_values * slope + intercept

equation = "y = " + str(round(slope,2)) + " x + " + str(round(intercept,2))

plt.scatter(x_values, y_values, marker = "o")
plt.plot(x_values,regress_values,"r-")

plt.title(f"Northern Hemisphere: Latitude vs Max Temperature ({real_date})")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.annotate(equation,(5,15),fontsize = 15, color="red")
plt.grid()

print(f"The r-value is : {rvalue**2}")
plt.savefig("WeatherPy_Outputs/Northern_Hemisphere_Max_Temp_vs._Latitude_Linear_Regression.png")
plt.show()
```
